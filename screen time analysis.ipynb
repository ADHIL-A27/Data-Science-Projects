```python
figure = px.bar(data_frame=data,
                x = "Date",
                y = "Usage",
                color = "App",
                title = "Usage"
                )
figure.show()
```

```python
figure = px.bar(data_frame=data,
                x = "Date",
                y = "Times opened",
                color = "App",
                title = "Times Opened"
                )
figure.show()
```

```python
figure = px.bar(data_frame=data,
                x = "Date",
                y = "Notifications",
                color = "App",
                title = "Times Opened"
                )
figure.show()
```

```python
figure = px.scatter(data_frame=data,
                    x = "Notifications",
                    y = "Usage",
                    size = "Notifications",
                    trendline= "ols",
                    title = "Relationship"
                    )
figure.show()
```
