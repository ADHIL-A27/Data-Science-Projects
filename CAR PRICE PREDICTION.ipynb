importing the dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

Data Collection and processing

```python
# loading the data from csv file to pandas dataframe
car_dataset =pd.read_csv("/content/car data.csv")
```

```python
#ispecting first five rows of the dataframe
car_dataset.head()
```

```python
# checking the number of rows and columns
car_dataset.shape
```

```python
#getting some infromation about the dataset
car_dataset.info()
```

```python
#checking the number of missing values
car_dataset.isnull().sum()
```

```python
# checking the distribution of categorical data
print(car_dataset.Fuel_Type.value_counts())
print(car_dataset.Seller_Type.value_counts())
print(car_dataset.Transmission.value_counts())
```

```python
#encoding "Fuel_Type" Coulmn
car_dataset.replace({'Fuel_Type':{'Petrol':0,"Diesel":1,"CNG":2}},inplace=True)

#encoding "Seller_Type" Coulmn
car_dataset.replace({'Seller_Type':{'Dealer':0,"Individual":1}},inplace=True)

#encoding "Transmission" Coulmn
car_dataset.replace({'Transmission':{'Manual':0,"Automatic":1}},inplace=True)
```

```python
car_dataset.head()
```

Spliting the data and Target


```python
X=car_dataset.drop(["Car_Name","Selling_Price"],axis=1)
Y=car_dataset['Selling_Price']
```

```python
print(X)
```

```python
print(Y)
```

Spliting Training and Test data


```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,random_state=2)
```

Model Training

1.Linear Regression


```python
# loading the linear regression model
lin_reg_model=LinearRegression()
```

```python
lin_reg_model.fit(X_train,Y_train)
```

Model Evaluation

```python
#prediction on Training data
training_data_prediction = lin_reg_model.predict(X_train)
```

```python
# R suared Error
error_score = metrics.r2_score(Y_train,training_data_prediction)
print("R squared Error:",error_score)
```

Visualize the actual prices and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Actual Prices vs Predicted Prices")
plt.show()
```

```python
#prediction on Training data
test_data_prediction = lin_reg_model.predict(X_test)
```

```python
# R suared Error
error_score = metrics.r2_score(Y_test,test_data_prediction)
print("R squared Error:",error_score)
```

```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Actual Prices vs Predicted Prices")
plt.show()
```

LASSO REGRESSION


```python
# loading the LASSO regression model
lass_reg_model=Lasso()
```

```python
lass_reg_model.fit(X_train,Y_train)
```

Model Evaluation

```python
#prediction on Training data
training_data_prediction = lass_reg_model.predict(X_train)
```

```python
# R suared Error
error_score = metrics.r2_score(Y_train,training_data_prediction)
print("R squared Error:",error_score)
```

Visualize the actual prices and predicted prices

```python
plt.scatter(Y_train,training_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Actual Prices vs Predicted Prices")
plt.show()
```

```python
#prediction on Training data
test_data_prediction = lass_reg_model.predict(X_test)
```

```python
# R suared Error
error_score = metrics.r2_score(Y_test,test_data_prediction)
print("R squared Error:",error_score)
```

```python
plt.scatter(Y_test,test_data_prediction)
plt.xlabel("Actual Price")
plt.ylabel("Actual Prices vs Predicted Prices")
plt.show()
```
